Saving Memory:

````python
before=id(Y)
Y=Y+X
id(Y)==before
````

````python
Z=torch.zeros_like(Y)
print(f'id(Z):{id(Z)}')
Z[:]=X+Y
print(f'id(Z):{id(Z)}')
````

### 2
Replace the two tensors that operate by element in the broadcasting mechanism with other shapes, e.g. 3-dimensional tensors. Is the result the same as expected?

````python
a=torch.arange(3).reshape((3,1,1))
b=torch.arange(6).reshape((1,2,3))
a,b,a+b
````

````python
a.shape,b.shape,(a+b).shape
````

Yes the results were the same as expected. Both tesnors were broadcasted into the lowest dimensions possible such that the elementwise addition operation would move forward.
