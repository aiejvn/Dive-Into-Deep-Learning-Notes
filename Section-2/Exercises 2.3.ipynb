1. Prove that the transpose of the tranpose of a matrix is the matrix itself: $(A^T)^T=A$.

````python
A=torch.arange(12).reshape((3,4))
A,A.T,A.T.T,(A==A.T.T).all()
````

Let 
$$
\displaystyle
A=[[a_{1,1}, a_{1,2}, \dots, a_{1,n}], 
    \dots, 
[a_{m,1}, a_{m,2}, \dots, a_{m,n}]],
A \in \mathbb{R}^{m \times n}
$$.

Then
$$
\displaystyle
A^T=[[a_{1,1}, a_{2,1}, \dots, a_{m,1}], 
    \dots, 
[a_{1,n}, a_{2,n}, \dots, a_{m,n}]],
A^T \in \mathbb{R}^{n \times m}
$$, and
$$
\displaystyle
{A^T}^T=[[a_{1,1}, a_{1,2}, \dots, a_{1,n}], 
    \dots, 
[a_{m,1}, a_{m,2}, \dots, a_{m,n}]],
{A^T}^T \in \mathbb{R}^{m \times n}
$$

$A^T$ is the same as ${A^T}^T$. QED.

2. Given two matrices $\bold{A}$ and $\bold{B}$, show that the sum and transposition commute: $\bold{A}^T + \bold{B}^T = (A+B)^T$

Let 
$$
\displaystyle
A=[[a_{1,1}, a_{1,2}, \dots, a_{1,n}], \newline
    \vdots, \newline
[a_{m,1}, a_{m,2}, \dots, a_{m,n}]], \newline
A \in \mathbb{R}^{m \times n}
$$.

Then
$$
\displaystyle
A^T=[[a_{1,1}, a_{2,1}, \dots, a_{m,1}], \newline
    \vdots, \newline
[a_{1,n}, a_{2,n}, \dots, a_{m,n}]], \newline
A^T \in \mathbb{R}^{n \times m}
$$. 

Similarly, let 
$$
\displaystyle
B=[[b_{1,1}, b_{1,2}, \dots, b_{1,n}], \newline
    \vdots, \newline 
[b_{m,1}, b_{m,2}, \dots, b_{m,n}]], \newline
B \in \mathbb{R}^{m \times n}
$$.

Then
$$
\displaystyle
B^T=[[b_{1,1}, b_{2,1}, \dots, b_{m,1}], \newline
    \vdots, \newline
[b_{1,n}, b_{2,n}, \dots, b_{m,n}]], \newline
B^T \in \mathbb{R}^{n \times m}
$$. 

$$
\displaystyle
A^T+B^T=[[a_{1,1}+b_{1,1}, a_{2,1}+b_{2,1}, \dots, a_{m,1}+b_{m,1}], \newline
    \vdots, \newline 
[a_{1,n}+b_{1,n}, a_{2,n}+b_{2,n}, \dots, a_{m,n}+b_{m,n}]] \newline
\newline

A+B=[[a_{1,1}+b_{1,1}, a_{1,2}+b_{1,2}, \dots, a_{1,n}+b_{1,n}], \newline

3. Given any square matrix $\bold{A}$, is $\bold{A}+\bold{A}^T$ always symmetric? Can you prove the reuslt by using only the reuslts of the previous two exercises?

From the previous two exercises, we know ${A^T}^T=A$ and $A^T+B^T=(A+B)^T$. 

$A+A^T$ is symmetric IFF $A+A^T == (A+A^T)^T$.

We can apply the second property to decompose $(A+A^T)^T$ into $A^T+{A^T}^T$.

$$
\displaystyle
(A+A^T)^T = A^T+{A^T}^T \newline
 = A^T+A (\text{By the first property we proved})\newline 
 = A+A^T (\text{Rearrange terms})\newline 
$$

Thus, $A+A^T$ is symmetric. QED.

Consider three large matrices, say $A \in \mathbb{R}^{2^{10} \times 2^{16}}$,$B \in \mathbb{R}^{2^{16} \times 2^5}$, $C \in \mathbb{R}^{2^5 \times 2^{14}}$, initialized with Gaussian random vairables. You want to compute the product $ABC$. Is there any difference in memory footprint and speed, depending on whether you compute $(AB)C$ or $A(BC)$. Why?

In a multiplication of M and N, with dimensions $m \times n$ and $n \times p$ respectively, the product is $m \times p$, and to compute each of its $mp$ entries, you need to multiply $n$ times and add. This means we have $mnp$ mutliplications and $mp$ summations.

$AB$ is $2^{10} \times 2^{5}$, so it requires $2^{10+16+5}=2^{31}$ multiplications and $2^{10+5}=2^{15}$ summations to calculate. $(AB)C$ is a further $2^{29}$ multiplications and $2^{24}$ summations, giving a total of $2^{31}+2^{29}$ multiplications and $2^{15}+2^{24}$ summations.

$BC$ is $2^{16} \times 2^{14}$, so it requires $2^{16+5+14}=2^{35}$ multiplications and $2^{16+14}=2^{30}$ summations to calculate. $A(BC)$ is a further $2^{40}$ multiplications and $2^{24}$ summations, giving a total of $2^{35}+2^{40}$ multiplications and $2^{30}+2^{24}$ summations.

$(AB)C$ requires more multiplications and summations to calculate than $A(BC)$. Thus, there is a difference in memory footprint and speed.

11. Consider three large matrices, say $A \in \mathbb{R}^{2^{10} \times 2^{16}}$,$B \in \mathbb{R}^{2^{16} \times 2^5}$, $C \in \mathbb{R}^{2^5 \times 2^{16}}$. Is there any difference in speed depending on whether you compute $AB$ or $AC^T$? Why? What changes if you initialize $C=B^T$ without cloning memory? Why?

In a multiplication of M and N, with dimensions $m \times n$ and $n \times p$ respectively, the product is $m \times p$, and to compute each of its $mp$ entries, you need to multiply $n$ times and add. This means we have $mnp$ mutliplications and $mp$ summations.

Both $AB$ and $AC^T$ are $2^{10} \times 2^{5}$, so both require $2^{10+16+5}=2^{31}$ multiplications and $2^{10+5}=2^{15}$ summations to calculate and thus should have no difference in speed depending on if $AB$ or $AC^T$ is calculated.

If we initialize $C=B^T$ without cloning memory, then $C$ is simply a lazy transpose view of $B$, where $C_{i,j}$ internally maps to $B_{j,i}$ without explicitly rearranging data in memory. While this saves space by not cloning $B$, if $B$ is stored in row-major order (i.e., row-by-row in memory), then $C$ will access its elements in column-major order, resulting in poor spatial locality and cache-unfriendly code. Thus, if we initialize $C=B^T$ without cloning memory, then $AC^T$ will be slower to compute than $AB$ due to $C$ being stored in a cache-unfriendly way.
