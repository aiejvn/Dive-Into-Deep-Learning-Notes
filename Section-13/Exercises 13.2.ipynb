```python
import os 
import subprocess
import numpy
import torch
from torch import nn
from d2l import torch as d2l
```

```python
# Warmup for GPU computation
device = d2l.try_gpu()
a = torch.randn(size=(1000,1000), device=device)
b = torch.mm(a,a)

with d2l.Benchmark('numpy'):
    for _ in range(10):
        a = numpy.random.normal(size=(1000,1000))
        b = numpy.dot(a,a)
        
# Using Pytorch is orders of magnitude faster.
with d2l.Benchmark('torch'):
    for _ in range(10):
        a = torch.randn(size=(1000,1000),device=device)
        b = torch.mm(a,a)
```

```python
# Again, but forcing operations to be executed synchronously.
with d2l.Benchmark('torch'):
    for _ in range(10):
        a = torch.randn(size=(1000,1000),device=device)
        b = torch.mm(a,a)
    # torch.cuda.synchronize(device) # No CUDA enabled
```
