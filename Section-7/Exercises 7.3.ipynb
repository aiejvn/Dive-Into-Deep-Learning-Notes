# 7.3: Padding and Stride

```python
import torch
from torch import nn
```

## 7.3.1: Padding

```python
# We define a helper function to calculate convolutions. It initializes the 
# convolutional layer weights and performs corresponding dimensionality
# elevations and reductions on the input and output
def comp_conv2d(conv2d, X):
    # (1,1) indicates that batch size and the number of channels are both 1
    X = X.reshape((1,1)+X.shape)
    Y = conv2d(X)
    # Strip the first two dimensions: examples and channels
    return Y.reshape(Y.shape[2:])

# 1 row and column is padded on either side, so a total of 2 rows or columns
# are added
conv2d = nn.LazyConv2d(1, kernel_size=3, padding=1) # note that total rows added = 1*2 = 2, 3-1=2 (recommended padding)
X = torch.rand(size=(8,8))
comp_conv2d(conv2d, X).shape
```

```python
# We use a convolution kernel with height 5 and width 3. The padding on either
# side of the height and width are 2 and 1, respectively
conv2d = nn.LazyConv2d(1, kernel_size=(5,3), padding=(2,1))
comp_conv2d(conv2d,X).shape
```

## 7.3.2: Stride

```python
conv2d = nn.LazyConv2d(1, kernel_size=3, padding=1, stride=2)
comp_conv2d(conv2d,X).shape
```

```python
conv2d = nn.LazyConv2d(1, kernel_size=(3,5), padding=(0,1), stride=(3,4))
comp_conv2d(conv2d, X).shape
```

# Summary

* Padding can increases height and width of the output; often used to give output same height and width as input to avoid undesirable shrinkage of the output
    * Also ensures all pixels are used equally frequently
* We typically use symmetric padding on both sides of input height and width ($(p_h,p_w)$ padding)
* Most commonly we set $p_h=p_w$, where we just state we choose padding $p$
* $s_h=s_w$ $\rightarrow$ stride $s$
* Stride reduces resolution of output 
* By default, padding is 0 and stride is 1
* Padding with 0's has significant computational advantages
* Padding is not necessarily with zeros; there are alternatives
