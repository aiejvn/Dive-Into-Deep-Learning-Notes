# 6.6: File I/O

```python
import torch
from torch import nn
from torch.nn import functional as F
```

## 6.6.1: Loading and Saving Tensors

```python
x = torch.arange(4)
torch.save(x, 'x-file')
```

```python
x2 = torch.load('x-file')
x2
```

```python
# storing a list of tensors and reading them into memory
y = torch.zeros(4)
torch.save([x,y], 'x-files')
x2, y2 = torch.load('x-files')
(x2, y2)
```

```python
# writing and reading a dictionary that maps strings to vectors; convenient if we want to read or write all weights in a model
mydict = {'x':x, 'y':y}
torch.save(mydict, 'mydict')
mydict2 = torch.load('mydict')
mydict2
```

## 6.6.2: Loading Model Parameters

```python
# Note: in order to reinstate a mode, we need to generate the architecture in code and then load parameters from disk.
# This is because models themselves can contain arbitrary code.
class MLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.LazyLinear(256)
        self.output = nn.LazyLinear(10)

    def forward(self, x):
        return self.output(F.relu(self.hidden(x)))
    
net = MLP()
X = torch.randn(size=(2,20))
Y = net(X)
```

```python
torch.save(net.state_dict(), 'mlp.params')
```

```python
# Recovering the model
clone = MLP() # clone original MLP model
clone.load_state_dict(torch.load('mlp.params'))
clone.eval()
```

```python
Y_clone = clone(X)
Y_clone == Y
```
