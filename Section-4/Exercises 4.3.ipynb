# 4.3: The Base Classification Model

```python
import torch
from d2l import torch as d2l
```

## 4.3.1: The Classifier Class

```python
class Classifier(d2l.Module): #@save
    """The base class of classification models."""
    def validation_step(self, batch):
        Y_hat = self(*batch[:-1])
        self.plot('loss', self.loss(Y_hat, batch[-1]), train=False)
        self.plot('acc', self.accuracy(Y_hat, batch[-1]), train=False)
```

```python
@d2l.add_to_class(d2l.Module) #@save
def configure_optimizers(self): # Configure Minibatch SGD optimizer as default
    return torch.optim.SGD(self.parameters(), lr=self.lr) 
```

## 4.3.2: Accuracy

```python
@d2l.add_to_class(Classifier)
def accuracy(self, Y_hat, Y, averaged=True):
    """Compute the number of correct predictions"""
    Y_hat = Y_hat.reshape((-1, Y_hat.shape[-1]))
    preds = Y_hat.argmax(axis=1).type(Y.dtype) # Cast as Y type, as Python == operator is sensitive to data types
    compare = (preds == Y.reshape(-1)).type(torch.float32)
    return compare.mean() if averaged else compare
```

# 4.3.3: Summary

- Classification is common enough of a problem to warrant convenience functions
- Central improtance is accuracy (among other objectives for statistical and computational reasons)
- Always useful to have a convenience method for assessing accuracy of classifier empirically
