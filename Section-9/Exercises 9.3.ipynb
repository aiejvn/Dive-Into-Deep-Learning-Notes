## 9.3.3: Partitioning Sequences

```python
@d2l.add_to_class(d2l.TimeMachine) #@save
def __init__(self, batch_size, num_steps, num_train=10000, num_val=5000):
    # n=5, d=2
    super(d2l.TimeMachine, self).__init__()
    self.save_hyperparameters()
    corpus, self.vocab = self.build(self._download())
    array = torch.tensor([corpus[i:i+num_steps+1]
                        for i in range(len(corpus) - num_steps)])
    self.X, self.Y = array[:,:-1], array[:,1:]
```

```python
@d2l.add_to_class(d2l.TimeMachine) #@save
def get_dataloader(self, train):
    idx = slice(0, self.num_train) if train else slice(
        self.num_train, self.num_train + self.num_val)
    return self.get_tensorloader([self.X, self.Y], train, idx)
```

```python
data = d2l.TimeMachine(batch_size=2, num_steps=10)
for X, Y in data.train_dataloader():
    # A minibatch of target sequencesc can be obtained by shifting the input sequences by one token.
    print('X:', X, '\nY:', Y) 
    break
```

# Summary

* Language models estimate joint probability of a text sequence
* For long sequences, $n$-grams provide a convenient model by truncating dependence
    * However, there is not enough frequency to deal efficienctly with infrequent word combinations via Laplace smoothing
* To train language models, we can randomly sample pairs of input sequences and target sequences in minibatches
* After training, we will use perplexity to measure language model quality
* Language models can be scaledup with increased data size, model size, and amount in training compute
