## 9.4.2: Recurrent Neural Networks with Hidden States

```python
# Proving that calculation of X_t W_{xh} + H_{t-1} W_{hh} is the same as the 
# matrix multiplication of the concatenation of X_t and H_{t-1} and the concatenation of
# W_{xh} and W_{hh}
X, W_xh = torch.randn(3,1), torch.randn(1,4)
H, W_hh = torch.randn(3,4), torch.randn(4,4)
torch.matmul(X, W_xh) + torch.matmul(H, W_hh)
```

```python
torch.matmul(torch.cat((X,H), 1), torch.cat((W_xh, W_hh), 0)) # Shapes of (3,5) and (5,4), respectively
```

# Summary
* RNN: Neural network that uses recurrent computation for hidden states
    * Hidden state can capture historical info of sequence up to current time step
    * Number of RNN model parameters does not grow as time increases
